# Task 1: Conceptual Understanding

### 1. What is the difference between "Love" and "love" in NLP?


### Answer:

In NLP, "Love" and "love" are considered different tokens if the text is case-sensitive.

- "Love" (capital L) may represent:
  - Beginning of a sentence
  - A proper noun (e.g., a name)

- "love" (lowercase) usually represents:
  - A general verb or noun (emotion)

Without normalization (like lowercasing), the model treats them as two separate words

### 2. What happens if stopwords are not removed?


### Answer:
Stopwords are common words such as "is", "the", "and", and "in".

If they are not removed:

- The dataset becomes large and noisy  
- Important words get less weight  
- Model training becomes slower  
- Accuracy may decrease in tasks like classification  

👉 Example:  
"The movie is very good"

### 3. Provide two real-world scenarios where removing stopwords can be harmful.

### Answer:
Removing stopwords can sometimes change the meaning of a sentence.

- **Sentiment Analysis:**  
  "I do not like this product" → "like product"  
  (Negative meaning becomes positive)

- **Question Answering / Search Systems:**  
  "What is the time now?" → "time now"  
  (Important context is lost)

### 4. Explain the difference between stemming and lemmatization.

### Answer:

  **Stemming:**
- It removes the ending of a word using simple rules, often without considering meaning. The result may not be a real word.

**Lemmatization:**
- It converts a word to its base (dictionary) form using proper language rules, so the result is always a meaningful word.

# Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    """
    Preprocess input text by applying multiple cleaning steps:
    - Lowercasing
    - Removing URLs & emails
    - Removing numbers
    - Handling repeated characters
    - Removing extra spaces
    - Removing very short words (<=2, except 'no', 'not')
    """
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Remove special characters (keep alphabets & spaces)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 8. Remove very short words (<=2), except "no", "not"
    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    # Join back
    text = " ".join(tokens)
    
    return text
```

```python
# Test sentences
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Run stress test
for i, sentence in enumerate(test_sentences, 1):
    cleaned_sentence = preprocess_text(sentence)
    tokens = cleaned_sentence.split()
    
    print(f"\n--- Test Case {i} ---")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
```

# Task 4: Token Analytics

```python
# Token Analytics without pandas

analysis_results = []

for i, sentence in enumerate(test_sentences, 1):
    cleaned = preprocess_text(sentence)
    tokens = cleaned.split()
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens > 0:
        avg_token_length = round(sum(len(word) for word in tokens) / total_tokens, 2)
    else:
        avg_token_length = 0
    
    # Store results
    analysis_results.append({
        "original": sentence,
        "cleaned": cleaned,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": avg_token_length
    })
    
    # Print nicely
    print(f"\n--- Sentence {i} ---")
    print("Original Text       :", sentence)
    print("Cleaned Sentence    :", cleaned)
    print("Total Tokens        :", total_tokens)
    print("Unique Tokens       :", unique_tokens)
    print("Average Token Length:", avg_token_length)
```

# Task 5: Frequency Analysis

```python
from collections import Counter

# Step 1: Collect all tokens
all_tokens = []

for sentence in test_sentences:
    cleaned = preprocess_text(sentence)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

# Step 2: Count frequency
word_freq = Counter(all_tokens)

# Step 3: Top 10 most frequent words
most_common = word_freq.most_common(10)

# Step 4: Top 5 least frequent words
least_common = word_freq.most_common()[:-6:-1]

# Display results
print("Top 10 Most Frequent Words:")
for word, count in most_common:
    print(f"{word} : {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in least_common:
    print(f"{word} : {count}")
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    """
    Processes a list of text inputs and returns:
    - All tokens combined
    - Cleaned sentences
    """
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        cleaned = preprocess_text(text)
        tokens = cleaned.split()
        
        clean_sentences.append(cleaned)
        all_tokens.extend(tokens)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("Cleaned Sentences:")
for s in result["clean_sentences"]:
    print("-", s)

print("\nAll Tokens:")
print(result["tokens"])
```

# Task 7: Error Handling

```python
def preprocess_text(text):
    """
    Clean and preprocess a single text input
    Handles edge cases like empty strings, emojis, and numbers
    """
    
    # ✅ Handle None or empty input
    if not text or not isinstance(text, str):
        return ""
    
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Remove special characters (including emojis)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # ✅ If nothing left after cleaning
    if not text:
        return ""
    
    # 8. Remove short words (<=2), except "no", "not"
    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    return " ".join(tokens)
```

## Test Edge Cases

```python
edge_cases = [
    "",                  # empty string
    "😍😍🔥🔥",          # only emojis
    "1234567890",        # only numbers
]

print("\n===== ERROR HANDLING TEST =====")

for i, text in enumerate(edge_cases, 1):
    result = preprocess_text(text)
    print(f"\nCase {i}: '{text}'")
    print("Output:", result if result else "[EMPTY STRING]")
```
