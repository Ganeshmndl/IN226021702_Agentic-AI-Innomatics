```python
# Predefined login credentials
stored_username = "admin"
stored_password = "1234"

# User input
username = input("Enter username: ")
password = input("Enter password: ")

# Login validation
if username == stored_username and password == stored_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Counters for pass and fail
pass_count = 0
fail_count = 0

# Analyzing results
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Displaying final result
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
# List of user names
names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
cleaned_names = []

# Cleaning and standardizing names
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Printing the cleaned list
print("Cleaned Names List:", cleaned_names)
```

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyzing message lengths
for message in messages:
    message_length = len(message)
    
    if message_length > 10:
        print(f"Message: '{message}' | Length: {message_length} | Status: Long Message")
    else:
        print(f"Message: '{message}' | Length: {message_length}")
```
