```python
# Taking input
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Removing duplicate words
unique_words = set(words)

# Printing result
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Employee salary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_paid_employee = max(employees, key=employees.get)
highest_salary = employees[highest_paid_employee]

# Printing result
print("Highest Salary:", highest_paid_employee, "-", highest_salary)
```

```python
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initializing max and min values
max_value = numbers[0]
min_value = numbers[0]

# Finding max and min  by using loop
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

# results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
# Attendance data
attendance = ["P", "P", "A", "P", "P"]

# Counting present days
present_count = 0
for status in attendance:
    if status == "P":
        present_count += 1

# Calculating percentage
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

# result
print("Attendance Percentage:", attendance_percentage)
```

```python
# List of phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Removing duplicates using set
unique_phone_numbers = set(phone_numbers)

# result
print("Unique phone numbers:", unique_phone_numbers)
```

```python
# Input string
text = "python"

# Dictionary to store character frequency
char_frequency = {}

# Counting character occurrences
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

# result
print(char_frequency)
```

```python
# Dictionary of employees
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key_to_check = "Ravi"

# Checking if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# List of student marks
marks = [70, 75, 80, 65]

# Calculating total and average
total_marks = 0
for mark in marks:
    total_marks += mark

average_marks = total_marks / len(marks)

# result
print("Average Marks:", average_marks)
```
