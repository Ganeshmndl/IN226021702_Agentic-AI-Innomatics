## Task 1: Replace Hardcoded Prompt using PromptTemplate

```python
# Import PromptTemplate
from langchain_core.prompts import PromptTemplate

# Define the template (no f-strings used)
template = "Explain {topic} in simple terms for beginners"

# Create PromptTemplate object
prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Generate output using the template
output = prompt_template.format(topic="Artificial Intelligence")

# Display result
print(output)
```

## Task 2: Multi-Input Prompt System

```python
from langchain_core.prompts import PromptTemplate

# Define template with multiple inputs
template = "Explain {topic} for {audience} in a {tone} tone"

# Create PromptTemplate object
multi_prompt_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Test Cases
output1 = multi_prompt_template.format(
    topic="AI",
    audience="beginners",
    tone="friendly"
)

output2 = multi_prompt_template.format(
    topic="Python",
    audience="kids",
    tone="fun"
)

output3 = multi_prompt_template.format(
    topic="Deep Learning",
    audience="engineers",
    tone="technical"
)

# Print outputs
print(output1)
print(output2)
print(output3)
```

## Task 3: Prompt Variations Engine

```python
from langchain_core.prompts import PromptTemplate

# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Input
topic = "Machine Learning"

# Generate outputs
teaching_output = teaching_template.format(topic=topic)
interview_output = interview_template.format(topic=topic)
story_output = story_template.format(topic=topic)

# Print outputs
print("Teaching Style:\n", teaching_output)
print("\nInterview Style:\n", interview_output)
print("\nStorytelling Style:\n", story_output)
```

## Task 4: ChatPromptTemplate System

```python
from langchain_core.prompts import ChatPromptTemplate

# Create ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

# Example input
role = "teacher"
topic = "Neural Networks"

# Generate messages
messages = chat_template.format_messages(
    role=role,
    topic=topic
)

# Print messages
for msg in messages:
    print(f"{type(msg).__name__}: {msg.content}")
```

## Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    """
    Validates audience and tone inputs.
    Raises ValueError if inputs are invalid.
    """

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        raise ValueError(f"Invalid audience: {audience}. Choose from {valid_audience}")

    if tone not in valid_tone:
        raise ValueError(f"Invalid tone: {tone}. Choose from {valid_tone}")

    return True
```

## ***Test:***

```python
# Valid Input Test
print("Valid Input Test:")
print(validate_inputs("beginner", "fun"))

print("\n----------------------\n")

# Invalid Input Test (Handled Properly)
print("Invalid Input Test:")

try:
    validate_inputs("kids", "fun")  # invalid audience
except ValueError as e:
    print("Error:", e)
```

## Task 6: Prompt Generator App

```python
from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):
    """
    Generates a dynamic prompt using PromptTemplate based on style.
    """

    # Step 1: Validate inputs
    validate_inputs(audience, tone)

    # Step 2: Define templates
    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    # Step 3: Check style
    if style not in templates:
        raise ValueError("Invalid style. Choose from teaching, interview, storytelling")

    # Step 4: Create PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    # Step 5: Generate output
    return prompt.format(topic=topic, audience=audience, tone=tone)
```
