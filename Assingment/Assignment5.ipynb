```python
# Function to analyze post engagement
def analyze_post_engagement(likes):
    total_likes = 0
    
    for like in likes:
        total_likes += like
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Sample input
likes = [200, 150, 300, 400]

# Function call
analyze_post_engagement(likes)
```

```python
# Function to check medicine stock
def check_medicine_stock(stock):
    print("Medicine Stock:", stock)
    
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")

# Sample input
medicine_stock = 6

# Function call
check_medicine_stock(medicine_stock)
```

```python
# Function to check rainfall adequacy
def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0
    
    for value in rainfall_data:
        total_rainfall += value
    
    average_rainfall = total_rainfall / len(rainfall_data)
    
    print("Average Rainfall:", int(average_rainfall))
    
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

# Sample input
rainfall = [70, 75, 68, 80, 67]
required_rainfall = 70

# Function call
check_rainfall(rainfall, required_rainfall)
```

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Sample input
usernames = ["user1", "user2", "user1", "user3"]

# Function call
detect_duplicate_accounts(usernames)
```

```python
# Function to filter premium crop prices
def filter_premium_crops(prices):
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)

# Sample input
crop_prices = [1500, 2500, 1800, 3200]

# Function call
filter_premium_crops(crop_prices)
```

```python
# Function to evaluate system health
def check_system_health(error_count):
    print("Error Count:", error_count)
    
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

# Sample input
errors = 7

# Function call
check_system_health(errors)
```

```python
# Function to check transaction limit
def check_transaction_limit(amount):
    daily_limit = 50000
    print("Transaction Amount:", amount)
    
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

# Sample input
transaction_amount = 60000

# Function call
check_transaction_limit(transaction_amount)
```

```python
# Function to check exam eligibility based on attendance
def check_exam_eligibility(attendance):
    present_days = 0
    
    for status in attendance:
        if status == "P":
            present_days += 1
    
    attendance_percentage = (present_days / len(attendance)) * 100
    
    print("Attendance Percentage:", attendance_percentage)
    
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

# Sample input
attendance = ["P", "P", "A", "P", "P"]

# Function call
check_exam_eligibility(attendance)
```

```python
# Function to calculate electricity bill and usage level
def calculate_electricity_bill(units):
    bill_amount = 0

    if units <= 100:
        bill_amount = units * 3
    elif units <= 200:
        bill_amount = (100 * 3) + ((units - 100) * 5)
    else:
        bill_amount = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill_amount < 500:
        usage = "Low Usage"
    elif bill_amount <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill_amount, usage

# Sample input
units_consumed = 250

# Function call
bill, usage_status = calculate_electricity_bill(units_consumed)

# Printing result
print("Total Bill Amount:", bill)
print("Usage Level:", usage_status)
```
